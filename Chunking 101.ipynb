Next we need to load the dataset into a Pandas Dataframe. At ~50 Gigabytes, it is far too large to fit into the machine, so we chunk it into much smaller pieces at 1 million rows per piece. You made have to manually install s3fs with the "conda install --name tensorflow_p36 s3fs" command in the terminal command prompt

```python
%%time
import io
import s3fs
import pandas as pd
data_c = pd.read_csv('s3://hackathon-dataset-boihackpart1-eu-west-1-tsys/dataset/hackathon_credit.txt', sep = '\t', encoding = 'UTF-16', iterator = True)
```

```python
data_c.get_chunk(10000000)
```

```python
%%time
data_c.get_chunk(100000000)['TransactionAmount'].describe()
```

```python
%%time
for chunk in data_c:
    data_c.get_chunk(chunk)['TransactionAmount'].round(0).astype(int64)
```

```python
%%time
TranMean = []
for chunk in data_c:
    TranMean.add(chunk['TransactionAmount'].mean())
```

Let's calculate the average transaction amount in the dataset. Because the data is split into chunks, we need to iterate through each chunk and calculate the mean for that chunk. We then append that mean to a list, and do a final calculation on that array. Note you will have to do this for all summary statistics of central tendency: calculate the metric on each chunk, then do a final calculation. 

```python
TranMean
```

```python
print(sum(TranMean)/len(TranMean))
```
